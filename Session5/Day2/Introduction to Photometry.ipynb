### 1.1 Centroiding: Center of Mass

Calculate Center of Mass (CoM) defined as: $\bar{x} = \frac{\sum A_i x_i}{\sum A_i }$, $\bar{y} = \frac{\sum A_i  y_i}{\sum A_i }$.

```python
def cent_weight(n):
    """
    Assigns centroid weights
    """
    wghts=np.zeros((n),np.float)
    for i in range(n):
        wghts[i]=float(i-n/2)+0.5
    return wghts

def calc_CoM(psf, weights):
    """
    Finds Center of Mass of image
    """
    cent=np.zeros((2),np.float)
    temp=sum(sum(psf)-min(sum(psf))) # just get the flux from the object. Get the sum in 1D - the background, and sum it all.
    
    cent[1]=sum((sum(psf) - min(sum(psf)))* weights)/temp
    cent[0]=sum((sum(psf.T) - min(sum(psf.T)))* weights)/temp
    return cent
```

Use centroiding algorithm to find the actual centers of the targe and comparison.

```python
## Cut a box between search limits, centered around targcen
targbox = partimg[targcen[0]-searchr : targcen[0]+searchr, targcen[1]-searchr : targcen[1]+searchr] #making a box that is offset around the center of the target
weights = cent_weight(len(targbox)) #weight the lenght of the box
tcenoffset = calc_CoM(targbox, weights) # get the CoM
print(tcenoffset) # tell me what the offset is from my inital guess
tcenter = targcen + tcenoffset
```

### 1.2 Aperture Photometry

#### Science Aperture

```python
def circle(npix, r1):
    """
    Builds a circle
    """
    pup=np.zeros((npix,npix),np.int)
    for i in range(npix):
        for j in range(npix):
            r=np.sqrt((float(i-npix/2)+0.5)**2+(float(j-npix/2)+0.5)**2)
            if r<=r1:
                pup[i,j]=1
    return pup
```

#### Sky annulus

```python
def annulus(npix, r_inner,r_outer=-1.):
    """
    Builds an annulus
    """
    pup=np.zeros((npix,npix),np.int)
    for i in range(npix):
        for j in range(npix):
            #### Fill in annulus form here ####
            r=np.sqrt((float(i-npix/2)+0.5)**2+(float(j-npix/2)+0.5)**2)
            if ((r<=r_outer)&(r>=r_inner)):
                pup[i,j]=1
    return pup
```

## 2. Photometry

### 2.1 Calculate SNR

Calculate Signal-to-Noise Ratio. CCD noise = sqrt(signal + background + dark current + read noise). Ignore dark current and read noise here. 

```python
def calcsnr(target, bg):
    signal = target - bg
    noise = np.sqrt(signal + bg)
    snr = signal / noise
    return snr, noise   
```

Sum all flux inside target and comparison apertures and divide by number of pixels to get average count per pixel.

```python
targc = np.sum(targaper) / np.sum(circmask) #target center
targbg= np.sum(targann) /  np.sum(annmask) # target's background
compc = np.sum(compaper) /  np.sum(circmask)
compbg= np.sum(compann) /  np.sum(annmask)
```

```python
snr, noise = calcsnr(targc, targbg)
print(snr)
```

```python
snr, noise = calcsnr(compc, compbg)
print(snr)
```

### 2.2 Optimize photometry aperture

```python
## Write code here that tries a range of photometry apertures and finds the best SNR ##
aperatures = np.arange(1,10,1)   ## aperture array of values between 1 and 10

snrold = 0 #set the baseline for the snr
for aper in aperatures:
    apermask = circle(searchr*2, aper)
    targc= np.sum(apermask*newtarg)/np.sum(apermask) #sum area in the aperature*target CoM location and divde out the sum of the aperature?
    snr, noise = calcsnr(targc, targbg)
    if snr > snrold:
        bestaper = aper
        snrold = snr
    
snr, noise = calcsnr(targc, targbg)  # calcuate the values for the target and it's backgorund
    
```

### 2.3 Calculate the target's magnitude and uncertainty


Given the comparison is of known magnitude of 19.4

```python
targc = circle(searchr*2, ap_r)*newtarg
targskyc = annulus(searchr*2, sky_inner, sky_outer)*newtarg
compc = circle(searchr*2, ap_r)*newcomp
compskyc = annulus(searchr*2, sky_inner, sky_outer)*newcomp

ratio = np.sum(compc)/np.sum(targc) #ratio of the flux between the comp and target
sigratio = ratio*np.sqrt((np.sum(targc)/np.sum(targskyc))**2+ (np.sum(compc)/np.sum(compskyc))**2)  #error 

deltamag = -2.5*np.log10(ratio)  #apparent mag
sigmamag = 2.5*sigratio/(ratio*np.log(10)) #unc apparent mag

refmag = 19.4
absmag= refmag-deltamag
print("Measured Magnitude = {:0.3f} Â± {:0.3f}".format(absmag, sigmamag))
```
